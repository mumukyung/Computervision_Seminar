```python
def read_img(file_path):
  img_arr = cv2.imread(file_path)
  return cv2.cvtColor(img_arr,cv2.COLOR_RGB2BGR)
  
```

```python
import random
img_arrs=[]
img_num = range(0,100)
for img in random.sample(img_num,3):
  img_arrs.append(read_img(dolphin_img_list[img]))
  img_arrs.append(read_img(shark_img_list[img]))
  img_arrs.append(read_img(whale_img_list[img]))
```
